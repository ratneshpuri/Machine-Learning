```python
bank_dummies=pd.get_dummies(bank,drop_first=True)
```

```python
bank_dummies.head()
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
bank_dummies.columns
```

```python
x=bank_dummies.drop('y_yes',axis=1)
```

```python
x.columns
```

```python
y=bank_dummies.y_yes
```

```python
from sklearn import preprocessing
from sklearn import linear_model
```

```python
scaler=preprocessing.MinMaxScaler()
```

```python
x_scaled= scaler.fit_transform(x)
```

```python
model=linear_model.LogisticRegression()
```

```python
skf=StratifiedKFold(n_splits=10,shuffle=True,random_state=1)
lst_acc_stratified=[]
```

```python
for train_index, test_index in skf.split(x, y): 
    x_train_fold, x_test_fold = x_scaled[train_index], x_scaled[test_index] 
    y_train_fold, y_test_fold = y[train_index], y[test_index] 
    model.fit(x_train_fold, y_train_fold) 
    lst_acc_stratified.append(model.score(x_test_fold, y_test_fold)) 
```

```python
from statistics import mean, stdev 
```

```python
print('List of possible accuracy:', lst_acc_stratified) 
print('\nMaximum Accuracy That can be obtained from this model is:', 
      max(lst_acc_stratified)*100, '%') 
print('\nMinimum Accuracy:', 
      min(lst_acc_stratified)*100, '%') 
print('\nOverall Accuracy:', 
      mean(lst_acc_stratified)*100, '%') 
print('\nStandard Deviation is:', stdev(lst_acc_stratified)) 
```
